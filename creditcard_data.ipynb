```python
def box_plot(data):
    j=1
    fig = plt.figure(figsize=(14,10))
    for i in data.columns:
        if(data[i].dtype!="O"):
            plt.subplot(5,5,j)
            sns.boxplot(x=i,data=data)
            plt.tight_layout()
            j+=1
    plt.show()
def hist_plot(data):
    fig = plt.figure(figsize=(14,10))
    for i in data.columns:
        if((data[i].dtype=="int64") or (data[i].dtype=="float64")):
            sns.displot(x=i,data=data)
            plt.tight_layout()
    plt.show()
```

```python
box_plot(data)
```

```python
hist_plot(data)
```

```python
for i in data.columns:
    if(data[i].dtype=='float64'):
        data[i] = data[i].round(3)
for i in data.columns:
    if(data[i].dtype!='O'):
        data[i] = data[i].replace(np.inf,0)
```

```python
from sklearn.cluster import KMeans
```

```python
X = data.drop(['CUST_ID','PURCHASES'],axis=1)
n = 10
wcss=[]
for i in range(1,n):
    model = KMeans(n_clusters=i)
    model.fit(X)
    wcss.append(model.inertia_)
```

```python
plt.plot(wcss,marker='o')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
model = KMeans(n_clusters=2)
label = model.fit_predict(X)
```

```python
print("silhouette_score : ",silhouette_score(data.drop(["CUST_ID","label","PURCHASES"],axis=1), label))
```
